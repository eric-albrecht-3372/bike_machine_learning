```python
import pandas as pd
import numpy as np
import os
import datetime
```

```python
# import the weather data
print(os.getcwd())
weather_hour_df = pd.read_csv("BikeDatasets/weather_data_long.csv")
```

```python
# Print out the head
weather_hour_df.head()
```

```python
# grab the datetime object
weather_hour_df["date_time"] = pd.to_datetime(weather_hour_df["Date"])
weather_hour_df["date_time_hour"] = pd.to_datetime(weather_hour_df["Time"])
time_dict = {"year":[], "month":[], "day":[], "hour":[]}

# Define methods for pulling out the desired parameters
def get_day_params(input_value):
    time_dict["year"].append(input_value.year)
    time_dict["month"].append(input_value.month)
    time_dict["day"].append(input_value.day)
    return input_value

def get_hour_params(input_value):
    time_dict["hour"].append(input_value.hour)
    
weather_hour_df["date_time"].apply(lambda l: get_day_params(l))
weather_hour_df["date_time_hour"].apply(lambda l: get_hour_params(l))

weather_hour_df.drop(["Date", "Time", "date_time", "date_time_hour"], axis=1, inplace=True)

for subdict in time_dict.keys():
    weather_hour_df[subdict] = time_dict[subdict]
```
